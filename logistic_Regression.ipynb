```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
data={
    "hours":[1,2,3,4],
    "result":[0,0,1,1]
}
df=pd.DataFrame(data)
df
x=df[["hours"]]
y=df["result"]
model=LogisticRegression()
model.fit(x,y)
new=pd.DataFrame({"hours":[1.5]})
result=model.predict(new)
if result[0]==1:
    print("pass")
else:
    print("fail")
print(model.score(x,y))
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
data={
    "hours":[1,2,3,4],
    "result":[False,False,True,True]
}
df=pd.DataFrame(data)
df
x=df[["hours"]]
y=df["result"]
model=LogisticRegression()
model.fit(x,y)
new=pd.DataFrame({"hours":[2.5,1.5,9.5]})
result=model.predict(new)
for hours , res in zip(new["hours"],result):
    if res:
        print(f"{hours}====>pass")
    else:
        print(f"{hours}====>fail")
```
